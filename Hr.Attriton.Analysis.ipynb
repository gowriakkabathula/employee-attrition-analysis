```python
# which age groups leaves more
plt.figure(figsize=(10,5))

sns.histplot( data = df,
              x='Age',
              hue='Attrition',
              multiple='stack',
              bins = 20)
plt.title("Age Distribution by Attririon")
plt.show()
```

```python
bins = [18,25,35,40,45,60]
labels = ['18-25','25-35','35-40','40-45','45-60']
df['AgeGroup']=pd.cut(df['Age'],bins=bins,labels=labels)
```

```python
df['AgeGroup']
```

```python
sns.countplot(data = df,
              x='AgeGroup',
              hue= 'Attrition')
plt.title("Attrition by Age Group")
plt.show()
#Employees aged 25-35 may leave more frequently.
```

```python
# Is Salary a Reason?
plt.figure(figsize=(8,5))
sns.boxplot(data = df, color='green',
            x='Attrition',
            y='MonthlyIncome'
           )
plt.title("Salary vs Attrition")
plt.show()
# Employees with lower salaries tend to leave more
```

```python
# Does Overtime Cause Attrition?
sns.countplot(
    data = df,
    x='OverTime',
    hue='Attrition')
plt.title("OverTime Impact on Attrition")
plt.show()
# Employees working overtime are more likely to leave
```

```python
# Work-life Balance Analysis
sns.countplot(data=df,
              x='WorkLifeBalance',
              hue='Attrition')
plt.title("Work-Life Balance vs Attrition")
plt.show()
#Poor Work-Life balance increases attrition
```

```python
# Promotion Delay Analysis
sns.boxplot( data = df,color='yellow',
             x='Attrition',
             y='YearsSinceLastPromotion')
plt.title("Promotion Delay vs Attrition")
plt.show()
# Employees without promotions for long periods are more likely chances to leave.
```
